## Gaussian Naive Bayes Algorithm:

## Import the standard libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Load the data

```python
data = pd.read_csv('breast-cancer.csv')
data.head()
```

```python
data.shape
```

## id column is of no use hence we drop it

```python
data = data.drop('id', axis = 1)
data.head()
```

```python
data['diagnosis'].value_counts(normalize = True) * 100
```

## Observations:

- 357 (63%) begnin cancer patients
- 212 (37%) malignant cancer patients

## Dependent variable - diagnosis is binary hence we should apply Guassian Naive Bayes

```python
dic = {'M' : 0, 'B' : 1}
data['diagnosis'] = data['diagnosis'].replace(dic)
data.head()
```

## Check for missing values

```python
data.isnull().sum()
```

## Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms
```

```python
data[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean']] = mms.fit_transform(data[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean']])
data.head()
```

```python
data[['radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst']] = mms.fit_transform(data[['radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst']])
data.head()
```

## Seperate X and y

```python
X = data.drop('diagnosis', axis = 1)
y = data['diagnosis']
```

## Divide the data into train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

## Apply Gaussian Naive Bayes on X_train and y_train data

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb
```

```python
gnb.fit(X_train,y_train)
```

## Perform predictions on X_test

```python
y_pred = gnb.predict(X_test)
y_pred
```

## Model Evaluation

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```
